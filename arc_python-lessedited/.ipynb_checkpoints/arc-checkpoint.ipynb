## ARC Experiments

```python
import numpy as np
import sys
sys.path.append('pymc')
import pymc.uncf as uf
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

### Functions

```python
def run(sigma, num_batches, histories_per_batch):
   
    point = (4.25, 4.25, 4.25)
    shape = (17, 17, 17)
    delta = (0.5, 0.5, 0.5)
    strength = 1E10
    
    mesh   = uf.Mesh(shape=shape, delta=delta, sigma=sigma)
    source = uf.Source(point=point, strength=strength)
    uncf   = uf.MCUNCF(source=source, mesh=mesh)

    uncf.Nb = num_batches
    uncf.Np = histories_per_batch

    # Calculate flux
    uncf.calc_flux()

    # Get the flux and relative error
    flux  = uncf.flux
    error = uncf.tally.relative_error()
    
    return flux, error
```

```python
def plot_flux_1D(flux):
    x = np.arange(0, 8.5, 0.5)
    plt.plot(x, flux[:,8,8], label="Flux")
    plt.title("Line plot of neutron flux")
    plt.xlabel("x [cm]")
    plt.ylabel("Neutron flux [n/cm$^2$-s]")
    plt.yscale("log")
```

```python
def plot_error_1D(error):
    x = np.arange(0, 8.5, 0.5)
    plt.plot(x, error[:,8,8], label=f"{batches} Batches")
    plt.title("Relative error in the neutron flux")
    plt.xlabel("x [cm]")
    plt.ylabel("Relative error")
    plt.legend(loc="upper right", bbox_to_anchor=(1.45, 0.8))
```

```python
def plot_flux_2D(flux):
    x = np.arange(0, 8.5, 0.5)
    data = flux[:,:,8]
    vmax = np.max(data)
    vmin = vmax/1E5
    plt.pcolormesh(x, x, data, norm=LogNorm(vmin, vmax))
    plt.title("Z-slice of neutron flux")
    plt.xlabel("x [cm]")
    plt.ylabel("y [cm]")
    cb = plt.colorbar()
    cb.set_label("neutron flux [n/cm$^2$-s]")
```

```python
def plot_error_2D(error):
    x = np.arange(0, 8.5, 0.5)
    plt.pcolormesh(x, x, error[:,:,8])
    plt.title("Z-slice of the relative error in the neutron flux")
    plt.xlabel("x [cm]")
    plt.ylabel("y [cm]")
    cb = plt.colorbar()
    cb.set_label("Relative error")
```

### Run the simulation

```python
x=2
sigma = 1
batches = 1
histories = 1000
for i in range(x):
    myflux1, myerr1 = run(sigma, batches, histories)
    plot_error_1D(myerr1)
    batches *= 5
```

### Plot results in 1D

```python
plot_flux_1D(myflux1)
```

```python
plot_error_1D(myerr1)
```

### Plot results in 2D

```python
plot_flux_2D(myflux1)
```

```python
plot_error_2D(myerr1)
myerr1[:, 8,  8]
```
